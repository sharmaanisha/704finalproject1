```python
def time_features(df):

    #Create time series features based on the time series index.

    #House = df.copy()
    Household_Energy['hour'] = Household_Energy.index.hour
    Household_Energy['dayofweek'] = Household_Energy.index.dayofweek
    Household_Energy['month'] = Household_Energy.index.month
    Household_Energy['year'] = Household_Energy.index.year
    Household_Energy['dayofmonth'] = Household_Energy.index.day
    Household_Energy['weekofyear'] = Household_Energy.index.isocalendar().week
    return df

Household_Energy = time_features(Household_Energy)
```

```python
# Set the date range for the split
start_date = '2016-01-11 17:00:00'
end_date = '2016-05-27 18:00:00'

#total amount of days in the dataset is 137, approximately 80% would be 110 days,
#we will do 110 days for train and 27 days for test

# Split the DataFrame based on the date range
trainTS = Household_Energy[(Household_Energy.index >= start_date) & (Household_Energy.index < '2016-04-30 17:00:00' )]
testTS = Household_Energy[(Household_Energy.index >= '2016-04-30 17:01:00' ) & (Household_Energy.index <= end_date)]

# Check the sizes of the resulting DataFrames
print("TrainTS size:", len(trainTS))
print("TestTS size:", len(testTS))
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
trainTS.plot(ax=ax, label='Training Set', title='Data Train/Test Split')
testTS.plot(ax=ax, label='Test Set')
ax.axvline('2016-05-27 18:00:00', color='black', ls='--')
ax.legend(['Training Set', 'Test Set'])
plt.show()
```

```python
#Used XGBoost for timeseries forecasting
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error

features = ['hour', 'dayofweek', 'month', 'year']
target_VAR = 'Appliances'

x_trainTS = trainTS[features]
y_trainTS = trainTS[target_VAR]
x_testTS = testTS[features]
y_testTS = testTS[target_VAR]

XGBreg = xgb.XGBRegressor(base_score=0.5, booster='gbtree',n_estimators=1000, early_stopping_rounds=50,  objective='reg:linear', max_depth=3,
     learning_rate=0.01)

XGBreg.fit(x_trainTS, y_trainTS,eval_set=[(x_trainTS, y_trainTS), (x_testTS, y_testTS)], verbose=100)
```

```python
#base_score=0.5: The initial prediction score of all instances, a global bias term.

#booster='gbtree': The type of boosting model used. 'gbtree' indicates a tree-based model.

#n_estimators=1000: The number of boosting rounds (trees) to be run during training.

#early_stopping_rounds=50: If the performance on the validation set doesn't improve for the specified number of rounds, the training process will stop early.

#objective='reg:linear': The learning task and objective function. 'reg:linear' signifies a regression task using linear regression.

#max_depth=3: The maximum depth of each tree in the ensemble.

#learning_rate=0.01: The step size shrinkage used in each update. It's a regularization parameter to prevent overfitting.
```

```python
feature_importance = pd.DataFrame(data=XGBreg.feature_importances_,
             index=XGBreg.feature_names_in_,
             columns=['importance'])
feature_importance.sort_values('importance').plot(kind='barh', title='House_Energy Time Feature Importance', color='green')
plt.show()
```

```python
#Hour seems to have the highest feature importance of all the time features
#Day if the week then has the highest feature importancde for energy usage
#Dayofyear has the next highest feature importance
```

```python
testTS['prediction'] = XGBreg.predict(x_testTS)
Household_Energy = Household_Energy.merge(testTS[['prediction']], how='left', left_index=True, right_index=True)
ax = Household_Energy[['Appliances']].plot(figsize=(15, 5))
testTS['prediction'].plot(ax=ax, style='.')
plt.legend(['True data', 'Predictions'])
ax.set_title('Raw Data vs Predictions')
plt.show()
```
